# Merging datasets of each region for their monthly, daily, hourly and state data

```python
import pandas as pd
```

## Monthly Averages 

#### read and merge

```python
n = pd.read_csv("N_brazil_weather.csv")
ne = pd.read_csv("NE_brazil_weather.csv")
se = pd.read_csv("SE_brazil_weather.csv")
s = pd.read_csv("S_brazil_weather.csv")
cw = pd.read_csv("CW_brazil_weather.csv")
```

```python
df = pd.concat([n, ne, se, s, cw], ignore_index=True)
```

#### checking data

```python
df.head()
```

```python
df.shape
```

```python
df.region.value_counts()
```

#### export to csv

```python
df.to_csv("monthly_averages_brazil.csv", index=False)
```
